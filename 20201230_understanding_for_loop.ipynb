# 반복문
- for: 끝이 정해져 있는 경우
- while: 끝이 명확하지 않은 경우

```python
names = ['박도준', '박서현', '이도윤', '김선호', '배수지']
ages = [30, 33, 28, 35, 29]
```

```python
print(list(range(10)))
```

```python
for i in range(len(names)):
    print(f'이름: {names[i]}, 나이: {ages[i]}')
```

```python
for i, j in zip(names, ages):
    print(f'이름: {i}, 나이: {j}')
```

```python
tmp = [1, 2, 3, 4, 5]
for i in range(len(tmp)):
    print(i)
```

# for문 응용

```python
from PIL import Image
import pandas as pd
import os

dir = './picture/'
fNames = os.listdir(dir)

for f in fNames:
    img = Image.open(dir + f)
    img = img.resize((500, 500))
    img.save('./img/' + f)
    
print('-----done-----')
```

```python
print(len(fNames))
```

```python
for f in fNames:
    print(f, end=' ')
```

```python
# 리스트형
import os
from PIL import Image

dic1=[]
fName = os.listdir('./picture/')

for i in range(len(fName)):
    img=Image.open('./picture/'+fName[i])
    dic1.append([fName[i], img.format, img.width, img.height])
    
print(dic1)
```

```python
listTmp = []

for i in dic1:
    listTmp2 = []
    for n, j in enumerate(i):
        if n == 0 or n == 2:
            listTmp2.append(j)
    listTmp.append(listTmp2)

print(listTmp)
```

```python
import pandas as pd

dfTmp = pd.DataFrame(listTmp, columns=['파일명', '너비'])
dfTmp.head()
```

# for문으로 이미지 합성

```python
from PIL import Image
import os

dir = './img/'
fNames = os.listdir(dir)

logo = Image.open('logo.png')
area = (0, 0, logo.width, logo.height)

for n in fNames:
    
    if '.jpg' in n:
        img = Image.open(dir + n)
        img.paste(logo, area)
        img.save(dir + 'logo/' + n)

print('------- done --------')
```
