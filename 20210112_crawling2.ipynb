```python
import requests
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen
import pandas as pd
```

```python
url = 'https://news.daum.net/'

html = requests.get(url)
soup = bs(html.content, 'html.parser')

# html = urlopen(url)
# soup = bs(html, 'html.parser')
```

```python
headlines = soup.find_all('ul', {'class':'list_headline'})

titles = []
for h in headlines:
    tmp = h.find_all('strong', {'class':'tit_g'})
    for t in tmp:
        try:
            tit = t.find('a', {'class':'link_txt'}).text
            press = t.find('span', {'class':'info_news'}).text
            titles.append([press, tit])
        except:
            continue
    
df = pd.DataFrame(titles, columns=['언론사', '제목'])
df
```

# 함수 만들기

```python
def naver_crawling(query, nums):
    import requests
    from bs4 import BeautifulSoup as bs
    import pandas as pd

    ls_news = []
    for num in range(1, nums, 10):
        url = 'https://search.naver.com/search.naver?where=news&sm=tab_jum&query=' + query + '&sm=tab_pge&sort=0&photo=0&field=0&reporter_article=&pd=0&ds=&de=&docid=&nso=so:r,p:all,a:all&mynews=0&start=' + str(num) + '&refresh_start=0'
        html = requests.get(url + query)
        soup = bs(html.content, 'html.parser')
        boxs = soup.find_all('li', class_ = 'bx')

        for b in boxs:
            try:
                press = b.find('a', class_ = 'info press').text.replace('언론사 선정', '')
                title = b.find('a', class_ = 'news_tit').text
                desc = b.find('div', class_ = 'news_dsc').text
                ls_news.append([press, title, desc])
            except:
                continue

    df = pd.DataFrame(ls_news, columns=['언론사', '제목', '내용'])
    return df
```

# 다음 뉴스 크롤링

```python
def daum_crawling(query, page):
    import requests
    from bs4 import BeautifulSoup as bs
    import pandas as pd

    ls_news = []
    for p in range(1, page+1):
        url = 'https://search.daum.net/search?w=news&q='+query+'&DA=PGD&spacing=0&p='+str(p)
        html = requests.get(url + query)
        soup = bs(html.content, 'html.parser')
        boxs = soup.find_all('div', class_ = 'wrap_cont')
        for b in boxs:
            try:
                press = b.find('span', class_ = 'f_nb date').text.split('|')[1]
                title = b.find('div', class_ = 'wrap_tit mg_tit').text
                desc = b.find('p', class_ = 'f_eb desc').text
                ls_news.append([press, title, desc])
            except:
                continue

    df = pd.DataFrame(ls_news, columns=['언론사', '제목', '내용'])
    return df
```

```python
df1 = daum_crawling('강남역', 3)

print(df1.shape)
display(df1.tail())
```

```python
df2 = daum_crawling('서울역', 3)

print(df2.shape)
display(df2.tail())
```

# 한국경제 링크 크롤링

```python
def link_to_list(query, page):
    import requests
    from bs4 import BeautifulSoup as bs
    import pandas as pd

    ls_news = []
    for p in range(1, page+1):
        url = 'https://search.hankyung.com/apps.frm/search.news?query='+query+'&mediaid_clust=HKPAPER,HKCOM&page='+str(p)
        html = requests.get(url + query)
        soup = bs(html.content, 'html.parser')
        boxs = soup.find_all('div', class_ = 'txt_wrap')
        for b in boxs:
            try:
                hyper = b.find('a', {'target':'_blank'})['href']
                ls_news.append(hyper)
            except:
                continue
                
    return ls_news
```

```python
links = link_to_list('신년사', 1)
links
```
