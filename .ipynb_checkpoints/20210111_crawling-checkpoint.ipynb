```python
from bs4 import BeautifulSoup as bs

html_str = """
<html>
    <body>
        <ul class="greet">
            <li>hello</li>
            <li>bye</li>
            <li>welcome</li>
        </ul>
        <ul class="reply">
            <li>ok</li>
            <li>no</li>
            <li>sure</li>
        </ul>
        <div>
            <ul>
                <li>open</li>
                <li>close</li>
            </ul>
        </div>
    </body>
</html>
"""

bs_html = bs(html_str)
```

# 실습

```python
import requests
from bs4 import BeautifulSoup as bs

url = 'https://www.naver.com/'
html = requests.get(url)
soup = bs(html.content, 'html.parser')
```

```python
div = soup.find('div', {'class':'theme_cont'})
div
```

```python
ls_theme = soup.find_all('div', {'class':'list_theme_wrap'})

items = []
for theme in ls_theme:
    tmp = theme.find_all('li', {'class':'theme_item'})
    items += tmp
    
print(len(items))
```

```python
item_len = [len(i) for i in ls_theme]
print(len(ls_theme))
print(item_len)
```

```python
print(items[0])
```

```python
ls_info = []

for item in items:
    category = item.find('em').text
    title = item.find('strong').text
    desc = item.find('p', {'class':'desc'}).text
    date = item.find('span', {'class':'date'}).text
    source = item.find('span', {'class':'source'}).text
    
    ls_info.append([category, title, desc, date, source])
    
print(len(ls_info))
print(ls_info)
```

```python
import pandas as pd

df = pd.DataFrame(ls_info, columns = ['카테고리', '타이틀', '설명', '날짜', '출처'])

display(df.head())
display(df.tail())
```

# 수업 코드

```python
ul_tmps=soup.find_all('ul', {'class':'list_theme'})

for i in range(len(ul_tmps)):
    a_tmps=ul_tmps[i].find_all('a', {'class':'theme_info'})
    data = []
    for at in a_tmps:
        gro=at.find('em').text
        tit=at.find('strong').text
        by=at.find('p').text
        date=at.find('span').text
        data.append([gro, tit, by, date])
    
df1=pd.DataFrame(data, columns=['그룹','제목','내용','기간'])
df1
```

# 네이버 검색 결과 크롤링

```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd

query = input('검색 할 단어를 입력하세요: ')
url = 'https://search.naver.com/search.naver?where=news&sm=tab_jum&query=' + query
html = requests.get(url)
soup = bs(html.content)

div = soup.find_all('li', {'class':'bx'})

ls_news = []
for d in div:
    try:
        press = d.find('a', {'class':'info press'}).text
        title = d.find('a', {'class':'news_tit'}).text
        dsc = d.find('div', {'class':'news_dsc'}).text
        ls_news.append([press, title, dsc])        
    except:
        continue


df = pd.DataFrame(ls_news, columns=['언론사', '제목', '내용'])
df
```

```python
def get_news(word):
    import requests
    from bs4 import BeautifulSoup as bs
    import pandas as pd

    query = word
    url = 'https://search.naver.com/search.naver?where=news&sm=tab_jum&query=' + query
    html = requests.get(url)
    soup = bs(html.content)

    div = soup.find_all('li', {'class':'bx'})

    ls_news = []
    for d in div:
        try:
            press = d.find('a', {'class':'info press'}).text
            title = d.find('a', {'class':'news_tit'}).text
            dsc = d.find('div', {'class':'news_dsc'}).text
            ls_news.append([press, title, dsc])        
        except:
            continue


    df = pd.DataFrame(ls_news, columns=['언론사', '제목', '내용'])
    return df
```

# 파일 저장하기

```python
def get_news(word):
    import requests
    from bs4 import BeautifulSoup as bs
    import pandas as pd

    query = word
    url = 'https://search.naver.com/search.naver?where=news&sm=tab_jum&query=' + query
    html = requests.get(url)
    soup = bs(html.content)

    div = soup.find_all('li', {'class':'bx'})

    text = open('./html/news_' + word + '.txt', 'w')
    for d in div:
        try:
            press = d.find('a', {'class':'info press'}).text
            title = d.find('a', {'class':'news_tit'}).text
            dsc = d.find('div', {'class':'news_dsc'}).text
            text.write(press + '♠' + title + '♠' + dsc + '\n')
        except:
            continue
    text.close()
```

```python
get_news('종로')
```

```python
# 한줄씩 읽어오기
f = open('./html/news_종로.txt', 'r')
while True:
    text = f.readline()
    if not text:
        break
    print(text)
f.close()
```
