# for문으로 pandas 읽어오기

```python
import os
import pandas as pd

dir = './py_data/skt_m/'
ls_dir = os.listdir(dir)

for i, file in enumerate(ls_dir):
    if i == 0:
        df = pd.read_csv(dir + file)
    else:
        tmp = pd.read_csv(dir + file)
        df = pd.concat([df, tmp])
        
print('------ done -------')
```

```python
pd.set_option('display.max_rows', 10)  # None -> show all rows
print(df.shape)
display(df[:10])
```

```python
import os
import pandas as pd

dir = './py_data/skt_m/'
ls_dir = os.listdir(dir)

df = pd.DataFrame()

for file in ls_dir:
    df_tmp = pd.read_csv(dir + file)
    df = pd.concat([df, df_tmp])
    
print('------ done -------')
```

```python
import os
import pandas as pd

dir = './py_data/skt_m/'
ls_dir = os.listdir(dir)

df = pd.DataFrame()

for file in ls_dir:
    if '19년' in file:
        break
    
    df_tmp = pd.read_csv(dir + file)
    df = pd.concat([df, df_tmp])
    
print('------ done -------')
```

```python
import os
import pandas as pd

dir = './py_data/skt_m_sample/'
ls_dir = os.listdir(dir)
ls_dir_2016 = [file for file in ls_dir if '19년' in file]

for file in ls_dir_2016:
    df = pd.read_csv(dir + file, encoding='utf-8')
    df.rename(columns = {'이용비율(%)' : '이용_비율(%)'}, inplace = True)
    df.to_csv(dir + file, index = None)
    
print('----- done ------')
```

```python
print(df.shape)
display(df.head())
display(df.tail())
```
