# 조선일보 - 다른 검색 경로로 실행

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup as bs
from tqdm import tqdm
import time
import urllib
import re
import sys

# 브라우저 열기
keyword = '김정은'
page = 3
url = 'https://www.chosun.com/nsearch/?query=&siteid=&sort=1&date_period=all&writer=&field=&emd_word=&expt_word=&opt_chk=false&app_check=0'
driver = webdriver.Chrome('./py_data/crawling/chromedriver.exe')
driver.get(url)
time.sleep(1)

# 검색어 입력
search = driver.find_element_by_xpath('//*[@id="main"]/form/div/input')
search.click()
search.send_keys(keyword)
search.send_keys(Keys.ENTER)
time.sleep(3)

# 스크롤 내리기
for _ in range(page):
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    driver.find_element_by_xpath('//*[@id="load-more-stories"]').click()
    time.sleep(3)

# url 가져오기
html = driver.page_source
soup = bs(html, 'html.parser')
div = soup.find_all('div', class_='story-card-wrapper')

ls_link = []
for d in div:
    href = d.find('a', {'target':'_blank'})['href']
    ls_link.append(href)
```

```python
ls_link = []
cnt = 0
for d in div:
    try:
        href = d.find('a', class_='text__link story-card__headline | box--margin-none text--black font--primary h3 text--left')['href']
        ls_link.append(href)
    except:
        cnt += 1
```
