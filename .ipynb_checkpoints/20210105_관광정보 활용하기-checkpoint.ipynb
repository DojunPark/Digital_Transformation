# 관광 정보 지도에 출력하기

```python
df1.head()
```

```python
df2 = df1.iloc[:, [0, 5, 6]]
df2.head()
```

```python
print(df2.iloc[0, 0])
print(df2.iloc[0, 1])
print(df2.iloc[0, 2])
```

```python
import folium

center = [36.320250311618175, 127.65680132989054]
location = [df2.iloc[0, 1], df2.iloc[0, 2]]

m = folium.Map(location=center, zoom_start=7)

folium.Marker(location=location, 
              icon=folium.Icon(color='red'), 
             popup=df2.iloc[0, 0]).add_to(m)
```

```python
import folium

center = [36.320250311618175, 127.65680132989054]
m = folium.Map(location=center, zoom_start=7)

for i, r in df2.iterrows():
    folium.Marker(location=[r[1], r[2]],
                  icon=folium.Icon(color='red'), 
                  popup=r[0]).add_to(m)
```

```python
m
```

```python
m.save('./py_data/fastivals.html')
```

# 시작일과 종료일을 입력 받아서 축제 지도에 출력하기

```python
df.info()
```

```python
import folium

center = [36.320250311618175, 127.65680132989054]
m = folium.Map(location=center, zoom_start=7)

start = input('시작일을 입력하세요: ')
end = input('종료일을 입력하세요: ')

df_tmp = df[(df['시작일'] >= start) & (df['종료일'] <= end)]
print(df_tmp.shape)

for i, r in df_tmp.iterrows():
    folium.Marker(location = [r[5], r[6]],
                 icon = folium.Icon(color='red'),
                 popup=r[0]).add_to(m)
```
