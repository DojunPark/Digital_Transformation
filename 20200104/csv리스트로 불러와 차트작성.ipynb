```python
import matplotlib.font_manager as fm

font_path = 'malgunsl.ttf' 
fontprop = fm.FontProperties(fname=font_path, size=18)



import csv
listData=[]
file=open('경찰인력.csv','r')
csvfile=csv.reader(file)

for item in csvfile:
    listData.append(item)


import matplotlib.pyplot as plt

y=[]


sel=3
for i in range(1,len(listData)):   #2번자료부터 마지막자료(행)
    
    y.append(int(listData[i][sel])) # 1열은 '서울'


#plt.axis([2005,2018,0,10000])
plt.title(listData[0][sel], fontproperties=fontprop)
plt.plot(y)

plt.show()
```

```python
# [미션]

# listData중에서 1번줄의 (2005년도) 
# 1열에서 마지막열까지 자료만 tmp라는 변수에 할당
#  tmp를 출력하면 아래와 같은 값이 출력되어야함.
# 24085	7714	4575	4461	0	0	1856	12659	0	3684	2866	5817	4489	7361	5698	5550	1277

for i in range(1,len(listData)):
    tmp=listData[i][1:len(listData[i])]
    plt.plot(y,'bo')
    plt.title(listData[i][0])
    plt.show()

```
