```python
# ********** https://financedata.github.io/posts/matplotlib-hangul-for-windows-anaconda.html
#%matplotlib inline  #주피터 노트북에서 매트플롯할때 사용하는 작업임.필수작업은 아니나 차트 출력을 주피터안에서 하겠다는 선언구문임
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.random.randint(-100, 100, 50).cumsum()  # 차트그릴 샘플 데이터를 무작위로 생성
data
```

```python
# 차트작성 경고메세지나올수 있음. 한글이 출력되지 않음.네모형태로 화면에 표시됨
plt.plot(range(50), data, 'r')
plt.title('가격변동 추이')
plt.ylabel('가격')
```

### 폰트를 지정하는 두가지 방법
#### matplotlib에서 폰트를 지정하여 출력하는 하는 대표적인 방법은 크기 두 가지
#### 텍스트 출력시 폰트 특성을 지정: FontProperties 속성으로 폰트 속성을 지정
##### 전역으로 지정: rcParams 을 사용하여 전역으로 설정값을 지정
##### https://financedata.github.io/posts/matplotlib-hangul-for-windows-anaconda.html

```python
import matplotlib.font_manager as fm

font_path = 'c:/windows/fonts/malgunsl.ttf' 
fontprop = fm.FontProperties(fname=font_path, size=18)

plt.ylabel('가격', fontproperties=fontprop)
plt.title('가격변동 추이', fontproperties=fontprop)
plt.plot(range(50), data, 'r')
plt.show()
```
