```python
import matplotlib.pyplot as plt # as 뒤는 별명이고 사용자 마음
%matplotlib inline              # notebook을 실행한 브라우저에서 바로 그림을 볼 수 있게 해주는 것

plt.plot([3, 20,40,50])
plt.ylabel('값')
plt.show()
```

```python
import matplotlib.font_manager as fm
font_path = 'malgunsl.ttf' 
fontprop = fm.FontProperties(fname=font_path, size=18)

import matplotlib.pyplot as plt


x=['1과정',1,2,5,20]
y=['나이',20,25,24,30]
plt.plot(x[1:5], y[1:5]) 
plt.ylabel(y[0],fontproperties=fontprop)
plt.xlabel(x[0],fontproperties=fontprop) 
#plt.axis([0, 10, 0, 50])
x_min=min(x[1:5]) ; x_max=max(x[1:5])
y_min=min(y[1:5]) ; y_max=max(y[1:5])

plt.axis([x_min,x_max,y_min,y_max])
plt.show()

# 이런 에러는
# 'list' object has no attribute 'y' 
```
