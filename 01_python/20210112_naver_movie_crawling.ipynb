# 네이버 영화 평점 크롤링

```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd

date = '20210111'
# url = 'https://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=cur&date=' + date
url = 'https://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=pnt&date=' + date

html = requests.get(url)
soup = bs(html.content, 'html.parser')

titles = soup.find_all('div', class_='tit5')
points = soup.find_all('td', class_='point')

ls_info = []
for title, point in zip(titles, points):
    tit = title.find('a').text
    link = title.find('a')['href']
    pnt = point.text
    ls_info.append([date, tit, link, pnt])
    
df = pd.DataFrame(ls_info, columns=['날짜', '제목', '사이트', '점수'])
df
```

# 실습 코드

```python
from datetime import date, timedelta
import requests
from bs4 import BeautifulSoup as bs
import sys

url = 'https://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=cur&date='

# date1 = input('기준 날짜를 입력하세요(예:20201201): ')
# period = int(input('크롤링 하고자 하는 기간을 입력해 주세요: '))
date1 = '20201231'
period = 10

start_date = date(int(date1[:4]), int(date1[4:6]), int(date1[6:8]))
timegap = timedelta(days=period)
start_date = start_date - timegap
range_date = pd.date_range(start_date, periods=period, freq='D')

ls_info = []
for today in range_date:
    html = requests.get(url+today.strftime('%Y%m%d')) 
    soup = bs(html.text, "html.parser")
    
    titles = soup.find_all('div', class_='tit5')
    points = soup.find_all('td', class_='point')

    for title, point in zip(titles, points):
        tit = title.find('a').text
        link = title.find('a')['href']
        pnt = point.text
        ls_info.append([today, tit, link, pnt])

df = pd.DataFrame(ls_info, columns=['날짜', '제목', '사이트', '점수'])
df
```

```python
df['점수'] = df['점수'].astype(float)
df.info()
```

```python
# 피벗 테이블(중복제거)
mov_unique = pd.pivot_table(df, index = ['제목'], aggfunc=sum)
mov_unique.head()
```
