```python
from wordcloud import WordCloud, STOPWORDS
from konlpy.tag import Okt
from collections import Counter
import matplotlib.pyplot as plt
import nltk

okt = Okt()

with open('./py_data/2021_대통령_신년사.txt', 'r', encoding='utf-8') as f:
    txt = f.read()

stop_words = ['가야', '갈수록', '결코', '그동안', '다른', '달이', '대한', '두운', '또한', '로부터', '만큼', '면치', '물이',
             '보이', '스루', '위해', '헛되', '우리']

words = okt.pos(txt)
ls_noun = [word[0] for word in words if (word[1] == 'Noun') & (len(word[0])>1) & (word[0] not in stop_words)]

counts = Counter(ls_noun)

wordcloud = WordCloud(font_path = 'C:\Windows\Fonts\HMKMRHD.TTF',
                      background_color = 'black', 
                      width = 1000, height = 1000,
                      max_words = 500,
                      max_font_size = 300)

wc = wordcloud.generate_from_frequencies(counts)
plt.figure(figsize = (12, 12))
plt.imshow(wc, interpolation = 'bilinear')
plt.axis('off')

plt.savefig('./py_data/wordcloud/신년사_워드클라우드.jpg')
plt.show()
```

```python
import pandas as pd

count_sorted = counts.most_common()
df = pd.DataFrame(count_sorted, columns=['단어', '빈도수'])
df.to_csv('./py_data/wordcloud/신년사_단어빈도수.csv', encoding='euc-kr')

df[:15]
```

```python
import matplotlib.font_manager as fm

df2 = df[:10]

font_path = 'C:\Windows\Fonts\HMKMRHD.TTF'
fontprop = fm.FontProperties(fname = font_path, size = 15).get_name()
# plt.rc('font', family = fontprop)
plt.rcParams["font.family"] = fontprop
plt.rcParams["figure.figsize"] = (10,10)
plt.rcParams["font.size"] = 20

explode = [0, 0, 0.5, 0, 0, 0, 0, 0, 0, 0]

plt.pie(df2['빈도수'], labels=df2['단어'], autopct = '%.1f%%',
       startangle = 120, counterclock = False, explode = explode)

plt.show()
```
