# install Konlpy

### step1:  JDK download
- JDK는 자바머신
- `jdk-14.0.1_windows-x64_bin.exe` 다운로드: 버젼 유의!!
- 파일을 관리자 권한으로 실행하여 설치

### step2: 환경변수 설정
- 내PC - 속성 - 시스템 - 고급 시스템 설정 - 환경변수 - 시스템변수 - 새로만들기 선택
- 시스템변수 - 새로만들기 - 변수명: JAVA_HOME, 변수값: `C:\Program Files\Java\jdk-14.0.1` 추가
- 시스템변수 - 새로만들기 - 변수명: CLASSPATH, 변수값: `%JAVA_HOME%\lib` 추가
- 시스템변수 - path - 편집 - `%JAVA_HOME%\bin` 추가

### step3: JPype download
- JPype는 파이썬에서 Konlpy를 실행하기 위해 설치하는 패키지
-  python 버전에 맞는 JPype 다운로드
- `JPype1-1.2.0-cp38-cp38m-win_amd64.whl` 혹은 `pip install jpype1==1.2.0`: 버전 유의!!
- 파일을 파이썬 실행 폴더로 옮기기
- prompt창에서 파일이 있는 폴더로 이동 후, `pip install JPype1-1.2.0-cp38-cp38m-win_amd64.whl`

### step4: install Konlpy
- `pip install Konlpy`

### step5: reboot
- 에러가 발생 시 환경변수가 제대로 설정되었고, 버전에 맞는 jdk와 jpype1이 설치되었는 지 확인하고 재부팅

### NLTK
- 교육용으로 개발된 자연어 분석 파이썬 패키지
- 코퍼스, 토큰 생성, 형태소 분석, 품사 태깅 기능을 제공

**용어 설명**
1. **코퍼스**: 문서의 집합
2. **토큰 분석**: 문자열을 토큰(문자열 단위)으로 나누는 작업
3. **형태소 분석**: 어근, 접두사, 접미사, 품사 등 다양한 언어적 속성을 파악하고 처리하는 작업
4. **품사 태깅**: 문장을 문법적인 기능이나 형태, 뜻에 따라 구분한 것 e.g., NNP(단수 고유명사), VB(동사), DT(관형사)

```python
import nltk

nltk.download('maxent_ne_chunker')
```

```python
from nltk import Text
from nltk.tokenize import RegexpTokenizer
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

plt.rcParams["figure.figsize"] = (14,4)

with open('c:/edu/py_data/test.txt') as f:
    txt = f.read()
    
print(txt)
```

```python
retokenize = RegexpTokenizer('[\w]+')   # 문장을 공백을 기준으로 분할
text = Text(retokenize.tokenize(txt))   # 빈도 분석
text
```

```python
# 빈도 출력
plt.title('Title')
text.plot()
plt.show()
```

```python
# 단어가 사용된 위치 시각화
text.dispersion_plot(['you', 'for', 'tried', 'free'])
```

# 영어 텍스트 분석

### 코퍼스

```python
import nltk
nltk.download("book", quiet=True)
from nltk.book import *
```

```python
emma_raw = nltk.corpus.gutenberg.raw("austen-emma.txt")
print(emma_raw[:1302])
```

### 토큰화

```python
# 예시 문장
print(emma_raw[:100])
```

```python
# word_tokenize
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize

print(word_tokenize(emma_raw[:100]))
```

```python
#RegexpTokenizer
from nltk.tokenize import RegexpTokenizer
retokenize = RegexpTokenizer('[\w]+')   # 문자 또는 숫자가 1개 이상인 경우 인식 (구두점, 기호 제외)

print(retokenize.tokenize(emma_raw[:100]))
```

### 품사 태깅

```python
from nltk.tag import pos_tag
ls_tagged = pos_tag(word_tokenize(emma_raw[:100]))

print(ls_tagged)
```

### 명사 추출

```python
ls_nouns = [t[0] for t in ls_tagged if t[1]=='NN']
print(ls_nouns)
```
