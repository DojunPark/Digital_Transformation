# 이미지 각도 변경
- pic1, pic2, pic3 이미지 각도를 90도 회전
- 기존 파일명 앞에 rot90_를 붙여서 저장 (예: rot90_pic1.jpg)

```python
from PIL import Image 

fileName = 'pic1'
tmp = Image.open('./picture/' + fileName + '.jpg')
angle = 90

tmpRotate = tmp.rotate(angle)
tmpRotate.save('./img/' + 'rot' + str(angle) + '_' + fileName + '.jpg')

print('========== END ============')
```

- 3개 파일 모두 90도 회전

```python
from PIL import Image 

fileNames = ['pic1', 'pic2', 'pic3']

for i in range(len(fileNames)):
    tmp = Image.open('./picture/' + fileNames[i] + '.jpg')
    angle = 90

    tmpRotate = tmp.rotate(angle)
    tmpRotate.save('./img/' + 'rot' + str(angle) + '_' + fileNames[i] + '.jpg')

print('========== END ============')
```

### 미션
이미지 속성값을 이용하여 각 파일의 파일명, 가로, 세로 길이를 출력하기
- `tmp.filename` 
- `tmp.format`
- `tmp.mode`
- `tmp.size`
- `tmp.width`
- `tmp.height`

```python
from PIL import Image

fileNames = ['pic1', 'pic2', 'pic3']

for i in range(len(fileNames)):
    tmp = Image.open('./picture/' + fileNames[i] + '.jpg')
    
    print('-------------------')
    print(f'{i}번째 파일입니다.')
    print('file name: ', tmp.filename)
    print('format: ', tmp.format)
    print('mode: ', tmp.mode)
    print('size: ', tmp.size)

print('\n')
print('========= END ==========')
```

```python
from PIL import Image

fileNames = ['pic1', 'pic2', 'pic3']

for i in range(len(fileNames)):
    tmp = Image.open('./picture/' + fileNames[i] + '.jpg')
    
    print('-------------------')
    print('{:.3f}번째 파일입니다.'.format(float(i)))
    print('file name: ', tmp.filename)
    print('format: ', tmp.format)
    print('mode: ', tmp.mode)
    print('size: ', tmp.size)
    break

print('\n')
print('========= END ==========')
```

```python
# resize 함수 만들기

def image_resize(pName, width, height):
    
    from PIL import Image
    tmp = Image.open('./picture/' + pName + '.jpg')
    tmpResize = tmp.resize((width, height))
    tmpResize.save('./img/' + pName + '(' + str(width) + 'x' + str(height) + ')' + '.jpg')
    
    print('saved')
```

```python
image_resize('pic1', 500, 300)
```

```python
# 함수 생성

def image_resize_all(dir, width, height):
    
    import os
    from PIL import Image
    
    fNames = os.listdir(dir)
    
    for name in fNames:
        tmp = Image.open(dir + name)
        tmpResize = tmp.resize((width, height))
        tmpResize.save('./img/' + name + '(' + str(width) + 'x' + str(height) + ')' + '.jpg')

    print('saved')
```

```python
dir = './picture/'
image_resize_all(dir, 500, 300)
```

# 이미지 파일 정보를 csv 파일로 내보내기

```python
#dictionary to csv

from PIL import Image
import pandas as pd
import os

dir = './picture/'
fileNames = os.listdir(dir)

listFilename = []
listFormat = []
listMode = [] 
listSize = []
listWidth = []
listHeight = []

for name in fileNames:
    tmp = Image.open(dir + name)
    listFilename.append(tmp.filename)
    listFormat.append(tmp.format)
    listMode.append(tmp.mode)
    listSize.append(tmp.size)
    listWidth.append(tmp.width)
    listHeight.append(tmp.height)

dictFiles = {'파일명' : listFilename,
            '포맷' : listFormat,
            '모드' : listMode,
            '사이즈' : listSize,
            '너비' : listWidth,
            '높이' : listHeight}

# to csv
dfFiles = pd.DataFrame(dictFiles)
dfFiles.to_csv('df_files.csv', index = False, encoding='euc-kr')

dfTmp = pd.read_csv('df_files.csv', encoding='euc-kr')
dfTmp.head()
```

```python
# list to csv

from PIL import Image
import os
import pandas as pd

fileNames = os.listdir('./picture/')

infos = []

for name in fileNames:
    tmp = Image.open('./picture/' + name)
    infos.append([name, tmp.format, tmp.mode, tmp.size, tmp.width, tmp.height])

df_infos = pd.DataFrame(infos, columns=['파일명', '포맷', '모드', '사이즈', '너비', '높이'])
df_infos.to_csv('df_infos.csv', index=False, encoding='euc-kr')
df_tmp = pd.read_csv('df_infos.csv', encoding='euc-kr')
df_tmp.head()
```
