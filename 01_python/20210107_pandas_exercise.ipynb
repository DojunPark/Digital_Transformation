```python
import pandas as pd
import os

fdir = '/mnt/c/edu/py_data/주유소/'
ls_file = os.listdir(fdir)

df = pd.DataFrame()
for f in ls_file:
    tmp = pd.read_excel(fdir + f, header=2)
    df = pd.concat([df, tmp], ignore_index=True)
    
df.to_csv(fdir + '서울특별시주유소.csv', encoding='euc-kr')
```

```python
import googlemaps

gmaps = googlemaps.Client(key=key)

ls_loc = []
len_loc = []
for i, r in df.iterrows():
    address = r[2]
    info = gmaps.geocode(address, language='ko')
    len_loc.append(len(info))
    
    lat = info[-1]['geometry']['location']['lat']
    lng = info[-1]['geometry']['location']['lng']
    
    ls_loc.append([lat, lng])
    
print(len(ls_loc))
print(max(len_loc))
```
