# 시도별 전출입 인구수 

```python
import pandas as pd

df = pd.read_excel('./py_data/data/시도별 전출입 인구수.xls')
```

```python
df.head()
```

```python
df = df.fillna(method='ffill')  # nan을 위의 값으로 변환
df.head()
```

```python
df_seoul = df[(df['전출지별'] == '서울특별시') & (df['전입지별'] != '서울특별시')]
df_seoul
```

```python
del df_seoul['전출지별']
df_seoul
```

```python
df_seoul = df_seoul.set_index('전입지별')
df_seoul
```

```python
df_seoul_gk = df_seoul.loc['경기도']
display(df_seoul_gk)
```

```python
df_seoul_gk.values
```

```python
df_seoul_gk.index
```

# 충청남도, 경상북도, 강원도

```python
ls_years = list(range(1970, 2018))
print(ls_years)
```

```python
df_seoul.columns = ls_years
```

```python
df1 = df_seoul.loc[['충청남도', '경상북도', '경기도', '제주특별자치도', '부산광역시'], ls_years]
df1
```

```python
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

rc('font', family='NanumBarunGothic')

fig = plt.figure(figsize = (20, 5))
ax = fig.add_subplot(1, 1, 1)

# axe 객체에 plot 함수로 그래프 출력
ax.plot(ls_years, df1.loc['충청남도',:], marker='o', markerfacecolor='green', 
        markersize=10, color='olive', linewidth=2, label='서울 -> 충남')
ax.plot(ls_years, df1.loc['경상북도',:], marker='o', markerfacecolor='blue', 
        markersize=10, color='skyblue', linewidth=2, label='서울 -> 경북')
ax.plot(ls_years, df1.loc['경기도',:], marker='o', markerfacecolor='red', 
        markersize=10, color='magenta', linewidth=2, label='서울 -> 경기')


plt.xticks(rotation = 45)  # 라벨 회전
plt.legend(loc='best', fontsize=13)   #범례 표시
plt.title('서울 인구 이동', fontsize=15)
plt.show()
```
