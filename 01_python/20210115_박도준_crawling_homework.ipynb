- https://www.tripadvisor.co.kr/Restaurant_Review-g294197-d1978666-Reviews-or10-Gusto_Taco-Seoul.html
- 위 사이트에 접속하여 음식점 리뷰 가져오기
- 가져올 항목: 작성일자, 제목, 내용, 평점

```python
from selenium import webdriver
from bs4 import BeautifulSoup as bs
import pandas as pd
from tqdm import tqdm
import time


# 정규표현식
regex = re.compile('.*ui_bubble_rating bubble.*')

# 브라우저 열기
driver = webdriver.Chrome('./py_data/crawling/chromedriver.exe')

# 페이지 반복
ls_info = []
for num in tqdm(range(0, 550, 10), desc='crawling reviews...'):

    # 페이지 이동
    url = 'https://www.tripadvisor.co.kr/Restaurant_Review-g294197-d1978666-Reviews-or'+str(num)+'-Gusto_Taco-Seoul.html'
    driver.get(url)
    time.sleep(1)
    
    # 더보기 클릭
    driver.find_element_by_xpath("//*[@class='taLnk ulBlueLinks']").click()
    time.sleep(1)
    
    # 정보 추출
    html = driver.page_source
    soup = bs(html, 'html.parser')
    div = soup.find_all('div', class_='reviewSelector')
    
    for d in div:
        rating = d.find('span', class_=regex)['class'][-1][-2:]
        date = d.find('span', class_='ratingDate').text.replace('에 작성. ', '')
        title = d.find('span', class_='noQuotes').text
        text = d.find('p', class_='partial_entry').text
        ls_info.append([rating, date, title, text])

# 데이터프레임 생성
df = pd.DataFrame(ls_info, columns=['평점', '날짜', '제목', '글'])
df.to_excel('./py_data/crawling/서울_구스토타코_리뷰.xlsx', encoding='euc-kr')
```
