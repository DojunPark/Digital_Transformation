```python
# Null 값이 존재하는지 확인
print(train_data.isnull().values.any())
```

```python
train_data = train_data.dropna(how = 'any') # Null 값이 존재하는 행 제거
print(train_data.isnull().values.any()) # Null 값이 존재하는지 확인
```

```python
# 정규 표현식을 통한 한글 외 문자 제거
import re
train_data['document'] = train_data['document'].str.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣 ]","")
```

```python
# 불용어 정의
stopwords = ['의','가','이','은','들','는','좀','잘','걍','과','도','를','으로','자','에','와','한','하다']
```

```python
# 형태소 분석기 OKT를 사용한 토큰화 작업 
okt = Okt()
tokenized_data = []
for sentence in train_data['document']:
    temp_X = okt.morphs(sentence, stem=True) # 토큰화
    temp_X = [word for word in temp_X if not word in stopwords] # 불용어 제거
    tokenized_data.append(temp_X)
```

```python
# 리뷰 길이 분포 확인
print('리뷰의 최대 길이 :',max(len(l) for l in tokenized_data))
print('리뷰의 평균 길이 :',sum(map(len, tokenized_data))/len(tokenized_data))
plt.hist([len(s) for s in tokenized_data], bins=50)
plt.xlabel('length of samples')
plt.ylabel('number of samples')
plt.show()
```

```python
from gensim.models import Word2Vec
model = Word2Vec(sentences = tokenized_data, size = 100, window = 5, min_count = 5, workers = 4, sg = 0)
```

```python
# 완성된 임베딩 매트릭스의 크기 확인
model.wv.vectors.shape
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from konlpy.tag import Okt
tagger = Okt()
```

```python
def get_noun(text):
    nouns = tagger.nouns(text)
    return [n for n in nouns if len(n) > 1]   # 두 글자 이상인 명사 추출
```

```python
# TF-IDF 행렬 구하기

cv = TfidfVectorizer(tokenizer = get_noun, max_features = 100)
tdm = cv.fit_transform(articles)

display(tdm.toarray()[2])
```

```python
okt = Okt()
results = []
lines = articles
words_all = []
```

```python
for line in lines:
    # 형태소 분석하기
    malist = okt.pos(line, norm=True, stem=True)
    r = []
    for word in malist:
        # 명사/동사/부사만 걸러내기 
        if word[1] in ['Noun','Verb','Adjective']:
            r.append(word[0])
            words_all.append(word[0])
    rl = (" ".join(r)).strip()
    results.append(rl)
    print(rl)
```

```python
from gensim.models import word2vec

yang_file = './py_data/word2vec/yang.model'
with open(yang_file, 'w', encoding='utf-8') as fp2:
    fp2.write('\n'.join(results))
```

```python
data = word2vec.LineSentence(yang_file)
model = word2vec.Word2Vec(data, size=200, window=10, hs=1, min_count=2, sg=1)
model.save('./py_data/word2vec/yang_w2v.model')
```

```python
def plot_2d_graph(vocabs, xs, ys):
    plt.figure(figsize=(15, 15))
    plt.scatter(xs, ys, marker = 'o')
    for i, v in enumerate(vocabs):
        plt.annotate(v, xy=(xs[i], ys[i]))
```
