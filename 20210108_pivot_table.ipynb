# 피벗 테이블
- `pd.pivot_table(df, index='그룹화할 열 이름')`: 기본 계산(평균)
- `pd.pivot_table(df, index=['열이름1', '열이름2'], aggfunc=[sum, count, ...])`
- `pd.pivot_table(df, index=['열이름1', '열이름2'], aggfunc=numpy.mean)`

```python
import pandas as pd

fdir = '/mnt/c/edu/py_data/주유소/'
df = pd.read_csv(fdir + '서울특별시주유소(구분할).csv', index_col=0, encoding='euc-kr')
df.head()
```

```python
pd.pivot_table(df, index=['구']).head()
```

```python
df.info()
```

```python
df = df[(df['휘발유']!='-')&(df['경유']!='-')]
```

```python
df = df.astype({'휘발유':'int64', '경유':'int64'})
```

```python
df.to_csv(fdir + '서울특별시주유소(구분할).csv', encoding='euc-kr')
```

```python
pd.pivot_table(df, index=['구'], values=['휘발유', '경유']).head()
```

```python
pd.pivot_table(df, index=['구'], values=['휘발유', '경유'], aggfunc = ['sum', 'count', 'mean']).head()
```

```python
import numpy as np

pd.pivot_table(df, index=['구'], values=['휘발유', '경유'], aggfunc = [np.mean, np.sum, np.max,np.min]).head()
```

```python
pd.pivot_table(df, index=['구', '상표'], values=['휘발유', '경유'], aggfunc = [np.mean, np.sum, np.max,np.min]).head(20)
```

```python
pd.pivot_table(df, index=['구', '상표', '셀프여부'], values=['휘발유', '경유'], aggfunc = [np.mean, np.sum, np.max,np.min]).head(20)
```

```python
tmp = pd.pivot_table(df, index=['구', '상표', '셀프여부'], values=['휘발유', '경유'], aggfunc = [np.mean, np.sum, np.max,np.min]).head(20)

display(tmp['mean'].head())
display(tmp['mean']['경유'].head())
```

```python
import folium

df1 = df.sort_values(by = '휘발유')
df_min = df1.head(10)
df_max = df1.tail(10)

center = [37.521846836377605, 126.98395957575897]
m = folium.Map(location = center, zoom_start = 11)

for i, r in df_min.iterrows():
    folium.Marker(location = [r[10], r[11]], 
                 icon = folium.Icon(color='blue'),
                 popup = r[1]).add_to(m)
    
for i, r in df_max.iterrows():
    folium.Marker(location = [r[10], r[11]], 
                 icon = folium.Icon(color='red'),
                 popup = r[1]).add_to(m)
```
