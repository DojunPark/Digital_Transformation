```
# 데이터 불러오기
import pandas as pd
data = pd.read_csv('price_data.csv')

data.head()
```

```
# 데이터 타입 변경
import numpy as np
xy = np.array(data, dtype=np.float32)
xy[:5, :]
```

```
# xy 분할
x_data = xy[:, 1:-1]
y_data = xy[:, [-1]]

display(x_data[:5])
display(y_data[:5])
```

```
# 모델 구축
import tensorflow as tf
from tensorflow.keras import layers, optimizers, Sequential

def create_model():
    model = Sequential([layers.Dense(50, activation='relu', input_shape=(4,)),
                        layers.Dense(20, activation='relu'),
                        layers.Dense(1)])
    model.compile(loss='mse', optimizer=optimizers.Adam(lr=0.0001))
    return model
```

```
model = create_model()
model.fit(x_data, y_data, epochs=5, batch_size=5)
```

```
# 학습 모델 저장1
# 모델의 모든 정보를 저장하기에 추후 model 구축이 필요하지 않음
model.save('./models/model.h5')
```

```
# 학습 모델 저장2
# 가중치만 저장하고자 한다면 체크포인트로 저장하여야 함
model.save_weights('./models/my_checkpoint')
```
