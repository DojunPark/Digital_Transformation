```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# font_path = './font/malgunsl.ttf'
font_path = './font/KoPubWorld Batang Medium.ttf'  # 원하는 글꼴 선택
fontprop = fm.FontProperties(fname=font_path, size=15)

x = ['과정', 1, 2, 5, 20]
y = ['나이', 20, 25, 24, 30]
plt.plot(x[1:5], y[1:5])
plt.xlabel(x[0], fontproperties=fontprop)
plt.ylabel(y[0], fontproperties=fontprop)


#plt.axis([0, 10, 0, 50])  # x축의 시작과 끝, y축의 시작과 끝
x_min = min(x[1:5]); x_max = max(x[1:5])
y_min = min(y[1:5]); y_max = max(y[1:5])
plt.axis([x_min, x_max, y_min, y_max])
plt.show()

print(f'x_min: {x_min}, x_max: {x_max}, y_min: {y_min}, y_max: {y_max}')
```

```python
import matplotlib.font_manager as fn
font_path = './Windows/Fonts/malgun.ttf'
fontprop = fn.FontProperties(fname=font_path)
fl = fontprop.get_name()
plt.rc('font', family=fl)

plt.figure(figsize=(10, 5))
plt.plot(df[['서울', '인천']], marker='o', linestyle=':')
plt.title('서울시 경찰 현황 변화', font_properties=fontprop)
plt.xlabel('인덱스', font_properties=fontprop)
plt.ylabel('인원수', font_properties=fontprop)
plt.legend([['서울', '인천']])
plt.show()
```
