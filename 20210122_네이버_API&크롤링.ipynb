# 책 리뷰 가져오기

```python
def get_href(link):
    base_url = 'https://book.naver.com'
    html = requests.get(link)
    soup = bs(html.content, 'html.parser')

    div = soup.find('div', class_='book_info')

    href = base_url + div.find('a', {'id':'txt_desc_point'})['href']
    num = int(re.sub('[^0-9]', '', div.find('a', {'id':'txt_desc_point'}).find_all('strong')[-1].text))
    page = (num // 10) + 1
    
    ls_href = []
    if num > 0:
        for i in range(1, page+1):
            href_tmp = href + '&page=' + str(i)
            html_tmp = requests.get(href_tmp)
            soup_tmp = bs(html_tmp.content, 'html.parser')
            reviews = soup_tmp.find('div', class_='review')
            div = reviews.find_all('dt')

            for d in div:
                tmp = d.find('a')['href']
                if 'https://' in tmp:
                    ls_href.append(tmp)
                else:
                    ls_href.append(base_url + tmp)
            time.sleep(1)

    return ls_href
```

```python
from tqdm import tqdm

cnt = 1
dict_link = {}
for link in tqdm(ls_links, 'saving...'):
    dict_link[f'num{cnt}'] = get_href(link)
    cnt += 1
    
print('-------- 완료 --------')
```

```python
import requests
from bs4 import BeautifulSoup as bs
from selenium import webdriver
from tqdm import tqdm
import time
import re

driver = webdriver.Chrome('./py_data/crawling/chromedriver.exe')

text = ''
for link in tqdm(ls_links, 'crawling...'):
    
    if 'blog.naver.com' in link:
        try:
            driver.get(link)
            time.sleep(1)
            driver.switch_to.frame('mainFrame')
            html = driver.page_source
            soup = bs(html, 'html.parser')
            container = soup.find('div', class_='se-main-container')
            tmp = re.sub('[^0-9가-힣a-zA-Z,. ]', '', container.text)
        except:
            container = soup.find('div', {'id':'postViewArea'})
            tmp = re.sub('[^0-9가-힣a-zA-Z,. ]', '', container.text)
        
    elif 'book.naver.com' in link:
        html = requests.get(link)
        time.sleep(1)
        soup = bs(html.content, 'html.parser')
        container = soup.find('div', class_='review')
        tmp = re.sub('[^0-9가-힣a-zA-Z,. ]', '', container.text)
    
    text += tmp

driver.close()
    
print('------- done --------')
```

# WordCloud

```python
from ckonlpy.tag import Twitter, Postprocessor

twitter = Twitter()
twitter.add_dictionary(['나동빈', '한빛미디어', '코딩테스트', '문제풀이', '코딩 테스트'], 'Noun')
replace = {'코딩 테스트':'코딩테스트'}
stopwords = {'부터', '까지', '영상', '어요', '위해', '이진', '도록', '통해', '때문', '라면', '매우', '모든', '요즘', '정말', '트랙', 
             '어보', '요하', '대해', '다가', '어진', '어가', '가기', '고자', '이나', '수도', '만큼', '해도', '아무', '래서', '성하', 
             '시오', '래그', '더보', '따라서', '려면', '보시', '여러', '어도'}
postprocessor = Postprocessor(twitter, replace = replace, stopwords = stopwords)
ls_pos = postprocessor.pos(text)
```

```python
ls_tmp = []
for word, pos in ls_pos:
    if (len(word) > 1) & (pos == 'Noun'):
        ls_tmp.append(word)
```

```python
len(ls_tmp)
```

```python
from collections import Counter

counts = Counter(ls_tmp)
counts = counts.most_common()
```

```python
counts
```

```python
stopwords = []
for word, cnt in counts:
    if cnt < 5:
        stopwords.append(word)
```

```python
ls_nouns = [word for word in ls_tmp if word not in stopwords]
```
