# Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 6, 0.1)
sin = np.sin(x)
cos = np.cos(x)
```

```python
print(x)
print(sin)
print(cos)
```

```python
plt.plot(x, sin)
plt.show()
```

```python
plt.plot(x, cos)
plt.show()
```

# 같이 그리기

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 6, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label = 'sin', linestyle='-.')
plt.plot(x, y2, label = 'cos', linestyle='--')
plt.xlabel('x')
plt.ylabel('y')
plt.title('sin, cos, tan')
plt.legend()
plt.show()
```
