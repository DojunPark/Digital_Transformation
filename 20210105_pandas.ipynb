# for문을 반복하며 데이터 병합

```python
pwd
```

```python
import os
import pandas as pd

fdir = '/mnt/c/edu/py_data/skt_m/'
flist = os.listdir(fdir)
print(flist)

df0 = pd.DataFrame()
for f in flist:
    df1 = pd.read_csv(fdir + f)
    df1.rename(columns = {'이용비율(%)':'이용_비율(%)'}, inplace=True)
    df0 = pd.concat([df0, df1], ignore_index=True)
    
df0.info()
```

```python
display(df0.head(3))
display(df0.tail(3))
```
